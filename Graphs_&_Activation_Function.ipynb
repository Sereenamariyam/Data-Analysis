```python
# LINEAR FUNCTION
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)
y = 2*x + 1

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=3)
plt.title('Linear Function: y = 2x + 1', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

```python

# QUADRATIC FUNCTION
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)
y = x**2

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'r-', linewidth=3)
plt.title('Quadratic Function: y = xÂ²', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

```python


# EXPONENTIAL FUNCTION
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)
y = np.exp(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'g-', linewidth=3)
plt.title('Exponential Function: y = eË£', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.ylim(0, 20)
plt.show()
```

```python
# SIGMOID FUNCTION
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)
y = 1 / (1 + np.exp(-x))

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'm-', linewidth=3)
plt.title('Sigmoid Function: y = 1/(1+eâ»Ë£)', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

```python
# ReLU ACTIVATION FUNCTION
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)
y = np.maximum(0, x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'orange', linewidth=3)
plt.title('ReLU Activation Function: y = max(0, x)', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

```python
# TANH ACTIVATION FUNCTION
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)
y = np.tanh(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'purple', linewidth=3)
plt.title('Tanh Activation Function: y = tanh(x)', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```
