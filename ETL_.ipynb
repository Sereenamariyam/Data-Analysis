```python
import pandas as  pd
import numpy as np
url = '/content/sales-data.csv.txt'
```

# **EXTRACTION**

```python
#extracting the above file

df_sales_data = pd.read_csv(url)
df_sales_data
```

# **TRANSFORMATION**

```python
# check data type of all columns

df_sales_data.info()
```

```python
# convert "orderdate"  type to datetime instead of object to perform operations

df_sales_data['OrderDate'] = pd.to_datetime(df_sales_data['OrderDate'])
df_sales_data.info()
```

```python
# Add Calculated Columns
# a) Total Price per Order :
# TotalPrice=Quantity×UnitPrice×(1−Discount)

df_sales_data['TotalPrice'] = df_sales_data['Quantity'] * df_sales_data['UnitPrice'] * (1 - df_sales_data['Discount'])
df_sales_data
```

```python
# to give a loyalty badge to the customer based on the "TotalPrice". create a definition give silver , gold and platinum badge

def assign_loyalty_badge(total_price):
    if total_price > 20.00:
        return 'Platinum'
    elif total_price > 10.00:
        return 'Gold'
    else:
        return 'Silver'

df_sales_data['LoyaltyBadge'] = df_sales_data['TotalPrice'].apply(assign_loyalty_badge)
df_sales_data
```

# **LOADING**

```python
#covert this into an excel file

df_sales_data.to_excel('sales_data.xlsx', index=False)
```

```python
#covert this into a csv file

df_sales_data.to_csv('sales_data.csv', index=False)
```

```python
#  push the sales data into sqlite3 and display sales table without using sql query

import sqlite3

# Connect to SQLite database (creates the database if it doesn't exist)
conn = sqlite3.connect('sales_database.db')

# Push the DataFrame to a SQLite table named 'sales'
df_sales_data.to_sql('sales', conn, if_exists='replace', index=False)


# Read the entire 'sales' table back into a DataFrame
df_sales_from_db = pd.read_sql_query("SELECT * from sales", conn)

# Display the DataFrame
df_sales_from_db
```
