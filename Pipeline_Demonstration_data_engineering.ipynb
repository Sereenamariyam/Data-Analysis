# Libraries

```python
#import important libraries
import pandas as pd
```

# Loading Data

```python
url = "https://tinyurl.com/49nm5ku9"  #while building dont give path explicitly, instead assign it to variable
#https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv - the actual url
df_covid = pd.read_csv(url)
df_covid.head()
```

# Preprocessing

```python
df_covid.isnull().sum()

```

```python
# 1. Filter rows where 'Country/Region' is India
'''df_covid_ind = df_covid[df_covid['Country/Region'] == 'India']
df_covid_ind.head()'''



def filter_india(df):
    df_india = df[df['Country/Region'] == 'India']
    return df_india
df_covid_ind = filter_india(df_covid)
df_covid_ind.head()
```

```python
#checking null values
'''df_covid_ind.isna().sum()'''
def drop_null_columns(df):
  df = df.drop(columns = ["Province/State"],axis=1)
  df = df.dropna(axis=1)
  return df
df_covid_ind_preprocessed =  drop_null_columns(df_covid_ind)
df_covid_ind_preprocessed.head()
```

```python
#dropping column axis = 0 is for rows and axos = 1 is for columns
df_covid_ind.drop(columns=['Province/State'], axis = 1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer

custom_pipeline = Pipeline([
    ('filter_india', FunctionTransformer(filter_india)),
    ('drop_null_columns', FunctionTransformer(drop_null_columns))
])
```

```python
#testing the pipeline
df_output = custom_pipeline.fit_transform(df_covid)
df_output.head()
```
