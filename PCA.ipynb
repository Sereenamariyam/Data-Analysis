### PCA

```python
from re import X
# split data into features and target
X=df_iris.drop('species',axis=1) # features
y=df_iris['species'] # target
X.head(3)
```

```python
pca=PCA()
X_new=pca.fit_transform(X) # gets the eige vectors of the 4 features
X_new
```

```python
pca.get_covariance()
```

```python
explained_variance = pca.explained_variance_ratio_
explained_variance
# the percentage of data retrieved from each of the features is given
```

```python
total=sum(explained_variance)
total
```

```python
plt.bar(range(1, len(explained_variance)+1),explained_variance)
plt.xlabel('Principal Components')
plt.ylabel('Explained variance ratio')
plt.show()
```

```python
pca_2 = PCA(n_components=2)
X_new_2 = pca_2.fit_transform(X)
explained_variance_2 = pca_2.explained_variance_ratio_
print("Cumulative sum of explained variance ratio:", sum(explained_variance_2))
```

```python
# new dataframe
# iris data has been transformed using only 2 principal components
df_iris_pca=pd.DataFrame(X_new_2,columns=['PC1','PC2'])
df_iris_pca['species']=y
df_iris_pca.head(3)
```
